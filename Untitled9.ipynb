```python
# Check the data type of 'Order Date' column
print(df['Order Date'].dtype)

# Display the first few values of 'Order Date' to visually inspect the format
print(df['Order Date'].head())
```

```python
# Standardize text columns: convert to title case (except for column names with product names)
text_columns = df.select_dtypes(include='object').columns
```

```python
# Convert text values to title case
df[text_columns] = df[text_columns].apply(lambda col: col.str.title())
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df.head()
```

```python
# Select columns with object dtype (assuming they are string columns)
string_columns = df.select_dtypes(include='object').columns

# Check the data types of string columns
print("Data types of string columns:")
print(df[string_columns].dtypes)

# Remove leading and trailing whitespace from string columns
for col in string_columns:
    df[col] = df[col].str.strip()

# Standardize text columns (excluding 'product_name') to title case
text_columns_to_standardize = [col for col in string_columns if col != 'product_name']
df[text_columns_to_standardize] = df[text_columns_to_standardize].apply(lambda col: col.str.title())

# Display the first few rows to see the changes
print("\nDataFrame after preprocessing string columns:")
display(df.head())
```

```python
df.to_csv("cleaned_sales.csv", index=False)
```
